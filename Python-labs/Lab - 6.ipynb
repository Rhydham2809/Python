### 02) WAP to merge two dictionaries given by user.

```python
def dictionary_maker(n):
  dictionary1 = {}
  for i in range(0,n):
      key = input(f"Please enter key {i+1} : ")
      value = input(f"Please enter its value : ")
      dictionary1[key] = value
  return dictionary1  

n = int(input("Please enter number of data in dictionary 1 : "))
dictionay1 = dictionary_maker(n) 
n = int(input("Please enter number of data in dictionary 2 : "))
dictionary2 = dictionary_maker(n) 

dictionary2.update(dictionay1)

print(dictionary2)

```

### 02) WAP to count frequency in list by dictionary.

```python
# L=["a","b","c","d","a","a","b","b"]
n = int(input("Please enter number of elements you want to enter : "))
L=[]
for i in range(n):
  L.append(input("Please enter the value : "))
  
d = {}
for i in L:
  if i in d:
    d[i]+=1
  else:
    d[i]=1  

print(d)    
```

### 03) WAP to remove all the duplicate words from the list using dictionary.

```python
n = int(input("Please enter number of elements you want to enter : "))
L=[]
for i in range(n):
  L.append(input("Please enter the value : "))
  
d = {}
for i in L:
  if i in d:
    L.remove(i)
  else:
    d[i]=1  

print(L)


# OR

# d = dict.fromkeys(L,1)
# L = list(d.keys())
# print(L)
```
