### 01) WAP to create Calculator module which defines functions like add, sub,mul and div. create another file that uses the Calculator module.

```python
import calculator

n1 = int(input("Please enter number 1 : "))
n2 = int(input("Please enter number 2 : "))

opertor = int(input("Please enter the number of the operator : \n 1.add 2.sub 3.mul 4.div 5.mod"))

match opertor:
        case 1:
            ans = calculator.add(n1,n2)
            print("The answer of addition is : ",ans)
        case 2:
            print("The answer of subtraction is : ",calculator.sub(n1,n2))
        case 3:
            print("The answer of multiplication is : ",calculator.mul(n1,n2))
        case 4:
            print("The answer of division is : ",calculator.div(n1,n2))
        case 5:
            print("The answer of module is : ",calculator.mod(n1,n2))                
```

### 04) WAP to demonstrate the use of the math module.

```python
import math

number = 3.4
n1 = 4
n2 = 16
L=[2,3,4,1]

#for ceiling value of number
print("ceil value = ",math.ceil(number))

#for floor value of number
print("floor value = ",math.floor(number))

#for factorial of number
print("factorial value of 5 = ",math.factorial(5))

#for sum of items in iterable
print("Sum of iteams in list = ",math.fsum(L))

#for gcd 
print("gcf value of numbers = ",math.gcd(n1,n2))

#checking if value of numbers is close or not
print("value of numbers is close = ",math.isclose(n1,n2))

```

### 03) WAP to generate 100 random lottery tickets and pick two lucky tickets from it as a winner.

```python
import random

L=[]

while len(L)!=100:
    L.append(random.randint(10001,100000))


print("Tickets : ",L)
print("Winners : ",random.choices(L,k=2))
```
