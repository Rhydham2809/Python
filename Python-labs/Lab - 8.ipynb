### 01) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
file = open("primenumbers.txt","w")
count = 0
for i in range(1,101):
    for j in range(2,(i//2)+1):
        if(i%j==0):
            count = 1
    if count==0:
        file.write(str(i)+"\n") 
    else:
        count = 0
file.close()

file = open("primenumbers.txt","r")
print(file.read())
```

### 03) WAP to encrypt a text file.

```python
inputData = input("Please enter a string : ")
n = int(input("Please enter a key to encrypt"))

file = open("words.txt","w")
for i in inputData:
    file.write(chr(ord(i)+n))
file.close()

file = open("words.txt","r")
print(file.read())
```
