### 04) WAP that defines a function which returns 1 if the number is prime otherwise return 0.

```python
def prime_number(n):
  for i in range(2,n//2):
    if n%i==0:
      return 0
    else:
      return 1  

n = int(input("Please enter number : ")) 
x = prime_number(n)
print("Non Prime number") if x==0 else print("Prime number")
```

### 05) Write a function called primes that takes an integer value as an argument and returns a list of all prime numbers up to that number.

```python
def primes(n):
  for i in range(2,n+1):
    count = 0
    for j in range(2,i//2+1):
      if i%j==0:
        count = 1
    if(count==0):
      print(i, end=" ")

n = int(input("Please enter number : ")) 
x = primes(n)

```

### 07) WAP to find the factorial of a given number using recursion.

```python
def factorial(n):
  if n==1 or n==0:
    return 1
  else:
    return n*factorial(n-1)   

x =  factorial(6)
x   
```

### 10) Write a python program to implement all function arguments type
Positional arguments <Br>
Default argument <Br>
Keyword arguments (named arguments) <Br>
Arbitrary arguments (variable-length arguments args and kwargs)

```python
# Default argument
def default_argument(n1,n2):
  return n1+n2

print("default argument = ",default_argument(2,3))

# Positional arguments
def Positional_arguments(n1,n2):
  return n1-n2

print("Positional arguments = ",Positional_arguments(n2=2,n1=3))
```

```python
# Arbitrary arguments
def arbitrary_arguments(n1,*n2):
  sum = n1
  for i in n2:
    sum += i
  return  sum

print("Arbitrary arguments = ",arbitrary_arguments(2,3,4,5,6))
```

### 01) WAP to calculate power of a number using recursion.

```python
def power_function(base,power):
  if power==1:
    return base 
  else:
    return base*power_function(base,power-1)  

x =  power_function(2,3)   
x
```

### 02) WAP to count digits of a number using recursion.

```python
count = 0
def count_digit(n):
  global count 
  count = count + 1
  if n<=0:
    return count-1
  else:
    return count_digit(n//10)

x = count_digit(200)    
x
```

### 03) WAP to reverse an integer number using recursion.

```python
reverse=0
def reverse_number(n):
    global reverse
    if n>0:
        reminder=n%10
        reverse=reverse*10+reminder
        reverse_number(n//10)
    return reverse
n = int(input("Enter a number : "))
print(f"Reverse no. of {n} is {reverse_number(n)}")
```

### 04) WAP to convert decimal number into binary using recursion.

```python
def decimalToBinary(n):
    if n==0:
        return 0
    return n%2+10*(decimalToBinary(n//2))
n = int(input("Enter a number : "))
print(f"Binary no. of {n} is {decimalToBinary(n)}")
```
