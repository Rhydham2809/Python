### 01) WAP to demonstrate the use of Pie chart.

```python
x=['20','20','20','20','20']
c=['r','g','b','c','m']
l=['red','green','blue','cyan','magenta']
e=[0,0,0,0.2,0]
plt.pie(x,colors=c,labels=l,explode=e,shadow=True,autopct='%1.1f%%')
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]

plt.plot(x,cxMarks,c='b',ls=":",lw=2,marker='o')
plt.plot(x,cyMarks,c='r',ls="--",lw=1,marker='>')
plt.xlabel('Roll no')
plt.ylabel('CPI')
plt.legend(['CX','CY'])
plt.annotate(xy=[8,1],text='Lowest CPI')
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline

x=[5,10,15,20,25]
y=[45,23,43,64,6]
l=["1st","2nd","3rd","4th","5th"]
c=["r","c","g","b","m"]
w=[0.7,0.8,0.6,0.9,1]
plt.bar(x,y,label=l,width=w,color=c)
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 

cpis = np.random.randint(1,10,100)
plt.hist(cpis,bins=10,histtype="stepfilled",align="mid",label="CPI hist")
plt.legend()
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
x=[20,34,45,23,56]
y=[2,4,6,8,10]
plt.barh(y,x)
for index, value in enumerate(y):
    plt.text(value, index,
             str(value))
plt.show()    
    
```

```python

import matplotlib.pyplot as plt
x = ["A", "B", "C", "D"]
y = [1, 2, 3.5, 4]
plt.barh(x, y)
 
for index, value in enumerate(y):
    plt.text(value, index,
             str(value))
 
plt.show()
```

### 03) WAP to Display an Image in Grayscale in Matplotlib.
