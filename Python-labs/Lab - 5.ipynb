### 08) Create a list from the specified start to end index of another list.

```python
a = int(input("Please enter the number of elements to be entered in List : "))
list1 = []
list2 = []

for i in range(0,a):
  list1.append(int(input(f"Please enter number {i+1}: ")))

start_index = int(input("Please enter the starting index : "))
end_index = int(input("Please enter the ending index : "))
if end_index == a or start_index == a:
  end_index = end_index-1
  start_index = start_index - 1

for i in range(start_index,end_index+1):
  list2.append(list1[i])

print(list2)  
```

### 01) WAP to count Even and Odd numbers in a List.

```python
a = int(input("Please enter the number of elements to be entered in List : "))
list1 = []
even_number = 0
for i in range(0,a):
  list1.append(int(input(f"Please enter number {i+1}: ")))
  if(list1[i]%2==0):
    even_number = even_number + 1
print(f"Even numbers = {even_number} \nOdd numbers = {a-even_number}")    
```
