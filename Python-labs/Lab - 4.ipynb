### 01) WAP to check given string is palindrome or not.

```python
string1 = input("Please ente the string : ")
reversedString = string1[-1:-len(string1)-1:-1]
if(string1==reversedString):
  print("Palindrome")
else:
  print("Not palindrome")  
```

### 01) WAP to find out duplicate characters in given string.

```python
string = input("Please enter the string : ")
count = 0
count2 = 0
duplicate_char = ""
no = 0
for x in range(0,len(string)):
  count = string.count(string[x])
  if(count>1):
    no = no + 1
    if duplicate_char.find(string[x])==-1:
        duplicate_char =   duplicate_char + " " + string[x]
        count2 = count2 + 1
print(f"The number of character repeated is {count2} \nThe duplicate characters is/are {duplicate_char} \nThe total number of times characters repeated are  {no}")          

```

### 03) WAP to find Maximum frequency character in String.

```python
string = input("Please enter the string : ")
count = 0 
max = 0
Highest_freq = ""
for x in string:
  count = string.count(x)
  if(max<=count):
    max = count
    if Highest_freq.find(x)==-1:
      Highest_freq =  x;
print(Highest_freq)    
print(max)    
```
